### Plot Presence

```python
selData = df[['SP_Transportation', 'SP_3D', 'SP_Realism', 'SP_MoveFreely', 'SP_Presence']].dropna()
frequencies = {}
for i in selData.columns:
    frequencies[i] = selData[i].value_counts()
```

```python
plotdata = pd.DataFrame(frequencies)
plotdata = plotdata.transpose()
plotdata2 = plotdata.div(plotdata.sum(axis=1), axis=0)*100
plotdata2
```
